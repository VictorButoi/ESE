```python
from segment_anything import SamAutomaticMaskGenerator, sam_model_registry
import matplotlib.pyplot as plt
import cv2
```

Initialize SAM module for prediction

```python
sam_checkpoint = "/storage/vbutoi/weights/SAM/sam_vit_h_4b8939.pth"
model_type = "vit_h"
sam = sam_model_registry["vit_h"](checkpoint=sam_checkpoint)
mask_generator = SamAutomaticMaskGenerator(sam)
```

Load image to run the segmentation on

```python
ade_root = '/storage/vbutoi/datasets/ade20k/ADE20K_2021_17_01/images/ADE/training/cultural/apse__indoor/'
image_path = f'{ade_root}/ADE_train_00001472.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.axis('off')
plt.show()
```

Get the masks

```python
masks = mask_generator.generate(image)
```
